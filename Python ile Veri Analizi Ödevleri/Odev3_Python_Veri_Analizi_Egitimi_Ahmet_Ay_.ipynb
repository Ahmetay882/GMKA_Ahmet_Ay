### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
eur_try_2021 = yf.download("EURTRY=X", start="2021-01-01", end="2021-12-31")
```

2. Veriyi özet metodları ile inceleyiniz

```python
eur_try_2021.head(10)
```

```python
eur_try_2021.info()
eur_try_2021.describe()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
eur_try_2021 = eur_try_2021 [['Adj Close']]
eur_try_2021.columns = ['Kapanış (TRY)']
eur_try_2021
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
eur_try_2021['Normalize 2021'] = eur_try_2021['Kapanış (TRY)'] / eur_try_2021.iloc[0,0]
eur_try_2021
```

### BIST30 Endeksine bakarak EUR/TRY kur değerininin 2021 yılında sene sonunun sene başına oranla 1,5 kat arttığını gözlemlemekteyim.

5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
fig_1 = px.line(eur_try_2021, y = 'Normalize 2021', 
        width = 1000, height = 500,
        title = 'BIST30 Endeksi 2021 - Normalize (1 Ocak 2021)',
        hover_data = {'Normalize 2021' : ':.2f', 'Kapanış (TRY)' : ':.2f'},
        labels = {'Date' : 'Tarih', 'Normalize' : 'BIST30 Endeksi (TRY)'})
fig_1.show()
```

```python
fig_2 = go.Figure()
fig_2.add_trace(go.Bar(x=eur_try_2021.index, y=eur_try_2021['Normalize 2021'], name='Normalize BIST30'))
fig_2.update_layout(title='BIST30 ve Normalize 2021',
                  xaxis_title='Tarih', xaxis_dtick='M1',
                  yaxis_title='Kapanış (TRY)')
fig_2.show()
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
B30 = yf.download('XU030', start='2022-01-01', end='2022-12-31') #2022 yılına ait BIST30 verileri
eur_try_2022 = yf.download('EURTRY=X', start='2022-01-01', end='2022-12-31') #2022 yılına ait EUR/TRY kur verileri
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
eur_try_2022 = eur_try_2022 [['Adj Close']]
eur_try_2022.columns = ['Kapanış (TRY)']
eur_try_2022
```

```python
eur_try_2022['Normalize 2022'] = eur_try_2022['Kapanış (TRY)'] / eur_try_2022.iloc[0,0]
eur_try_2022
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
veri_2021 = pd.DataFrame()
veri_2021 = pd.concat([veri_2021, eur_try_2021], axis=1)

veri_2022 = pd.DataFrame()
veri_2022 = pd.concat([veri_2022, eur_try_2022], axis=1)
```

```python
total_veri = pd.concat([veri_2021, veri_2022], axis=0) 
total_veri
```

```python
total_veri
```

```python
total_normalize = pd.DataFrame()
total_normalize = pd.concat([eur_try_2021, eur_try_2022], axis=1) 
total_normalize
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
fig_3 = px.line(total_veri, y=['Normalize 2021', 'Normalize 2022'])
fig_3.show()
```

```python
fig_4 = go.Figure()
fig_4.add_trace(go.Bar(x=total_veri.index, y=total_veri['Normalize 2021'], name='Normalize BIST30'))
fig_4.add_trace(go.Scatter(x=total_veri.index, y=total_veri['Normalize 2022'], name='Normalize EUR/TRY'))

fig_4.update_layout(title='Normalize BIST30 ve Normalize EUR/TRY',
                  xaxis_title='Tarih', xaxis_dtick='M1',
                  yaxis_title='Kapanış (TRY)')
fig_4.show()
```
