```python
!pip install yfinance
!pip install plotly
import yfinance as yf
import plotly.express as px 
import plotly.graph_objects as go
import plotly.subplots as sp
import pandas as pd
import matplotlib.pyplot as plt
eur_try_2021 = yf.download("EURTRY=X", start="2021-01-01", end="2021-12-31")
eur_try_2021 = eur_try_2021 [['Adj Close']]
eur_try_2021.columns = ['Kapanış (TRY)']
eur_try_2021['Normalize 2021'] = eur_try_2021['Kapanış (TRY)'] / eur_try_2021.iloc[0,0]

B30 = yf.download('XU030', start='2022-01-01', end='2022-12-31')

eur_try_2022 = yf.download('EURTRY=X', start='2022-01-01', end='2022-12-31') 
eur_try_2022 = eur_try_2022 [['Adj Close']]
eur_try_2022.columns = ['Kapanış (TRY)']
eur_try_2022['Normalize 2022'] = eur_try_2022['Kapanış (TRY)'] / eur_try_2022.iloc[0,0]
```

```python
veri_2021 = pd.DataFrame()
veri_2021 = pd.concat([veri_2021, eur_try_2021], axis=1)

veri_2022 = pd.DataFrame()
veri_2022 = pd.concat([veri_2022, eur_try_2022], axis=1)

total_veri = pd.concat([veri_2021, veri_2022], axis=0) 
total_veri
```

```python
total_veri['değişim'] = total_veri['Kapanış (TRY)'].diff(1)
total_veri['değişim_yönü'] = total_veri['değişim'].map(lambda x: 'düşen gün' if x<0 else 'yükselen gün')
total_veri['değişim%'] = (total_veri['değişim'] / total_veri['Kapanış (TRY)'].shift(1))
total_veri
```

### Veriyi özetleme

1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/TRY kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
total_veri['ay_yıl'] = total_veri.index.strftime('%Y-%m')
total_veri
```

```python
veri_as = total_veri.pivot_table(values=['Kapanış (TRY)'], index='ay_yıl', aggfunc='last')
veri_as
```

2. Ay kırılımlı olarak ürettiğiniz tabloyu grafik olarak inceleyiniz.


```python
veri_as['değişim'] = veri_as['Kapanış (TRY)'].diff(1)
veri_as['değişim_yönü'] = veri_as['değişim'].map(lambda x: 'düşen gün' if x<0 else 'yükselen gün')
veri_as['değişim%'] = (veri_as['değişim'] / veri_as['Kapanış (TRY)'].shift(1))
veri_as
```

```python
fig_1 = px.bar(veri_as, y = 'değişim%',
       color = 'değişim_yönü', color_discrete_map={'düşen gün':'red', 'yükselen gün':'green'},
       text_auto='.0%')
fig_1.show()
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
veri_as2 = total_veri.resample('BM', convention='end').asfreq()
veri_as2
```

4. Resample yöntemi ile ürettiğiniz tabloyu grafik olarak inceleyiniz. 2. adımda elde ettiğiniz grafik ile karşılaştırarak yorumlayınız.

```python
fig_2 = px.bar(veri_as2, y = 'değişim%',
       color = 'değişim_yönü', color_discrete_map={'düşen gün':'red', 'yükselen gün':'green'},
       text_auto='.0%')
fig_2.show()
```
